```python
import numpy as np
import tensorflow as tf
import glob
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
import datetime

%matplotlib inline

def plot_train_val(name, train_loss, val_loss, ax):
    # Make the plot easier to read by thinning it
    n = len(train_loss)
    
    train_thinning = int(n / 200)
    
    if train_thinning == 0:
        train_thinning = 1
    
    x_train = np.arange(len(train_loss))[0::train_thinning]
    x_val = np.arange(len(val_loss))
    
    if len(x_val) > 0:
        x_val = np.interp(x_val, (0, x_val.max()), (0, len(train_loss)))

    ax.set_yscale('linear')
    #ax.set_ylim(15000, 50000)
    
    ax.plot(x_train, train_loss[0::train_thinning])
    ax.plot(x_val, val_loss)
    ax.set_title("%s, minimum train loss: %f" % (name, np.min(train_loss)))
    epochs = len(train_loss) * 10 / (131070 / 128)
    losses_per_epoch = int(len(train_loss) / epochs)
    #print("%s\ttrain\t%d" % (name, int(np.mean(train_loss[-1]))))
    print("%s\ttrain\t%d" % (name, int(np.mean(train_loss[-losses_per_epoch:-1])/10)))
    print("%s\tval\t%d" % (name, int(val_loss[-1]/10)))
    
model_dir = "../../workspace/eval/exp-models/"
loss_plots = [
    ["0", model_dir + "0"],
    ["1a", model_dir + "1a"],
    ["1c", model_dir + "1c"],
    ["1d", model_dir + "1d"],
    ["1e", model_dir + "1e"],
    ["1f", model_dir + "1f"],
    ["1g", model_dir + "1g"],
    ["1h", model_dir + "1h"],
    ["1i", model_dir + "1i"],
    
    ["2a", model_dir + "2a"],
    ["2b", model_dir + "2b"],
    ["2d", model_dir + "2d"],
    
    ["3a", model_dir + "3a"],
    ["3b", model_dir + "3b"],
    ["3c", model_dir + "3c"],
    ["3d", model_dir + "3d"],
    
    ["4a", model_dir + "4a"],
    ["4a", model_dir + "4a"],
    ["4b", model_dir + "4b"],
    ["4c", model_dir + "4c"],
    
    ["5a", model_dir + "5a"],
    ["5b", model_dir + "5b"],
    ["5c", model_dir + "5c"],
]

fig = plt.figure(figsize=(20,8))
for i, (name, path) in enumerate(loss_plots):
    ax = fig.add_subplot(len(loss_plots) // 3 + 1, 3, i + 1)
    plot_train_val(name, np.load(path + "/summary/train_loss.npy"), np.load(path + "/summary/val_loss.npy"), ax)
    
    events = list(tf.train.summary_iterator(glob.glob(path + "/summary/events.out.tfevents.*")[0]))
    duration = datetime.datetime.fromtimestamp(int(events[-1].wall_time)) - datetime.datetime.fromtimestamp(int(events[0].wall_time))
    print("%s\tdur\t%s" % (name, str(duration)))

plt.tight_layout()
```
